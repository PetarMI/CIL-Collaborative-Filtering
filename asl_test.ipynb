```python
from Collabrative_Filtering import collabrative_filtering
from helper import write_submission, csv_parse, submit_results
import pandas as pd
import scipy.sparse as sparse
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_data():
    td = pd.read_csv('./data/bootstrap_train.csv')
    
    user_ids = td['row_id'] - 1
    item_ids = td['col_id'] - 1

    R = sparse.coo_matrix((td['Prediction'], (user_ids, item_ids))).todense()
    
    return R, user_ids, item_ids
```

```python
R, uids, iids = read_data()
```

```python
print(R.shape)
```

```python
# Initialize weight matrix
W = R>0.5
W[W == True] = 1
W[W == False] = 0
# To be consistent with our Q matrix
W = W.astype(np.float64, copy=False)

R = np.asarray(R)
W = np.asarray(W)
```

```python
lambda_ = 0.1
n_factors = 100
m, n = R.shape
n_iterations = 2

X = 5 * np.random.rand(m, n_factors) 
Y = 5 * np.random.rand(n_factors, n)
```

```python
def get_error(Q, X, Y, W):
    return np.sum((W * (Q - np.dot(X, Y)))**2)
```

```python
def als():
    weighted_errors = []
    
    for ii in range(n_iterations):
        for u, Wu in enumerate(W):
            if(u % 100 == 0):
                print(u)
            X[u] = np.linalg.solve(np.dot(Y, np.dot(np.diag(Wu), Y.T)) + lambda_ * np.eye(n_factors),
                                   np.dot(Y, np.dot(np.diag(Wu), R[u].T))).T
        for i, Wi in enumerate(W.T):
            if(i % 50 == 0):
                print(i)
            Y[:,i] = np.linalg.solve(np.dot(X.T, np.dot(np.diag(Wi), X)) + lambda_ * np.eye(n_factors),
                                     np.dot(X.T, np.dot(np.diag(Wi), R[:, i])))
        #weighted_errors.append(get_error(Q, X, Y, W))
        print('{}th iteration is completed'.format(ii))
    
    weighted_Q_hat = np.dot(X,Y)
    
    return weighted_Q_hat

R_pred = als()
```
